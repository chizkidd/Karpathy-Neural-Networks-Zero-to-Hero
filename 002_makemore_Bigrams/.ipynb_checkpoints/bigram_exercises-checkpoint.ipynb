````python
t = {}
for w in words:
    #chs = ['<S>'] + list(w) + ['<E>']
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        #print(ch1, ch2, ch3)
        trigram = (ch1, ch2, ch3)
        #print(trigram)
        
        # add up all bigrams & count how often they occur
        # if bigram is not in the dictionary, b.get(bigram,0) returns 0
        t[trigram] = t.get(trigram, 0) + 1
````

## function for making n-grams for visualization input

````python
def extract_ngrams(names:list, n:int):
    names = ['.' + name + '.' for name in names] # to keep track of start and end of names
    ngrams_list = []
    for name in names:
        for i in range(len(name)-n+1):
            ngrams_list.append(name[i:i+n])
    return ngrams_list

def find_ngrams(names, n):
    #names = ['.' + name + '.' for name in names] 
    names = '.'.join(names)+'.'
    return list(zip(*[names[i:] for i in range(n)]))


def make_char_ngram(names:list, N:int):
    data = defaultdict(int) # for speed
    text = '.'+'.'.join(names)+'.'
    for i in range(len(text)-N+1):
        x = text[i:i+N] # actual N-gram
        data[x] += 1    # add 1 to this N-gram key
    return data
````

````python
N = torch.zeros((27, 27, 27), dtype=torch.int32)
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        N[ix1, ix2, ix3] +=1
        
P = (N+1).float() 
P /= P.sum(2, keepdim=True) 
````

# b. Neural Network Approach

----
```
                                                                       
    +------|------+    W0 [27, 27]     +-------|-------+               +------|------+    
    |  2 Inputs   |   -------------->  | Hidden        |               | Output      |    
    |             |    W1 [27, 27]     | Layer         |   -------->   |             |    
    | 2x27 inputs |   -------------->  | 2x27 neurons  |               | 27 outputs  |    
    +------|------+     W [54, 27]     +-------|-------+               +------|------+    
           |                                   |                              |          
           |                                   |                              |          
           X [196113, 54]                     W.X                             y
                                (no bias, no non-linearity)
```  



````python
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        #print(ch1, ch2, ch3)
        xs.append((ix1, ix2))
        ys.append(ix3)
        
xs = torch.tensor(xs) 
ys = torch.tensor(ys)

g = torch.Generator().manual_seed(1478)
W = torch.randn((27*2, 27), generator=g, requires_grad=True)

xenc = F.one_hot(xs, num_classes=27).float()  # input to the network: one-hot encoding
logits = xenc.view(-1, 27*2) @ W              # predict log-counts
counts = logits.exp()                         # counts, equivalent to N
probs = counts / counts.sum(1, keepdim=True)
#print(xenc.shape, counts.shape, probs.shape)
#loss = -probs[torch.arange(ys.shape[0]), ys].log().mean() # ys.shape[0] is the number of examples (allows us to avoid using a count variable)
````

````python
num = 5
nlls = torch.zeros(num)
for i in range(num):
    # i-th trigram:
    x = xs.numpy().tolist()[i] 
    y = ys[i].item() 
    #print(x)
    #print(y)
    print('--------\n')
    print(f'trigram example {i+1}: {itos[x[0]]}{itos[x[1]]}{itos[y]} (indexes {x},{y})')
    print('input to the neural net:', x)
    print('output probabilities from the neural net:', probs[i])
    print('label (actual next character):', y)
    p = probs[i, y]
    print(f'probability assigned by the net to the the correct character: {p.item():.6f}')
    logp = torch.log(p)
    #print('log likelihood:', logp.item()%.4f)
    print(f'log likelihood: {logp.item():.4f}')

    nll = -logp
    print(f'negative log likelihood: {nll.item():.4f}')
    nlls[i] = nll

print('\n=========')
print(f'average negative log likelihood, i.e. loss = {nlls.mean().item():.4f}')
````

## combining it all together
```
                                                                       
    +------|------+    W0 [27, 27]     +-------|-------+               +------|------+    
    |  2 Inputs   |   -------------->  | Hidden        |               | Output      |    
    |             |    W1 [27, 27]     | Layer         |   -------->   |             |    
    | 2x27 inputs |   -------------->  | 2x27 neurons  |               | 27 outputs  |    
    +------|------+     W [54, 27]     +-------|-------+               +------|------+    
           |                                   |                              |          
           |                                   |                              |          
           X [196113, 54]                     W.X                             y
                                (no bias, no non-linearity)
```  


---
We need to make sure the matrix size aligns for matrix multiplication.
* **W.X** = `xenc.view(-1, 27x2)` @ `W` ------>  **[196113, 54] x [54, 27]**

````python
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        #print(ch1, ch2, ch3)
        xs.append((ix1, ix2))
        ys.append(ix3)
        
xs = torch.tensor(xs) 
ys = torch.tensor(ys)
num = ys.nelement()
print('number of examples: ', num)

# initialize the 'network'
g = torch.Generator().manual_seed(1478)
W = torch.randn((27*2, 27), generator=g, requires_grad=True)
# Got stuck a bit on this part and didn't realize how many neurons I'd be needing, 
# started with (27x27, 27) but I think this was not necessary
# You just need 27x2, the first 27 will activate for the first input, the second 27 for the second one
# Double the inputs, double the neurons, that makes sense! 
````

````python
%%time
n_iter = 250
lr = 50
C = 0.01
losses = []
# gradient descent
for k in range(n_iter):

    # forward pass
    xenc = F.one_hot(xs, num_classes=27).float()       # input to the network: one-hot encoding
    logits = xenc.view(-1, 27*2) @ W                   # predict log-counts
    counts = logits.exp()                              # counts, equivalent to N
    probs = counts / counts.sum(1, keepdim=True)      # probabilities for next character
    loss = -probs[torch.arange(num), ys].log().mean() + C*(W**2).mean() 
    #print(loss.item())
    losses.append(loss)

    # backward pass
    W.grad = None # set to zero the gradient
    loss.backward()

    # update
    W.data += -lr * W.grad
print(f'\naverage negative log_likelihood = {loss.item():.4f}\n')
````

````python
print(f"train_loss (final): {losses[-1]:.4f}")
plt.plot(losses)
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("Loss on training data")
plt.grid() 
plt.plot()
````

---

````python
xs_tr, ys_tr = [], []

for w in train:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        xs_tr.append((ix1, ix2))
        ys_tr.append(ix3)
        
xs_tr = torch.tensor(xs_tr) 
ys_tr = torch.tensor(ys_tr)
num_tr = ys_tr.nelement()
print('number of examples: ', num_tr)

# initialize the 'network'
g = torch.Generator().manual_seed(1478)
W = torch.randn((27*2, 27), generator=g, requires_grad=True)
````

````python
%%time
n_iter = 250
lr = 50
C = 0.01
losses_tr = []
# gradient descent
for k in range(n_iter):

    # forward pass
    xenc_tr = F.one_hot(xs_tr, num_classes=27).float()      
    logits = xenc_tr.view(-1, 27*2) @ W               
    counts = logits.exp()                             
    probs = counts / counts.sum(1, keepdim=True)      
    loss_tr = -probs[torch.arange(num_tr), ys_tr].log().mean() + C*(W**2).mean() 
    #print(loss.item())
    losses_tr.append(loss_tr)

    # backward pass
    W.grad = None # set to zero the gradient
    loss_tr.backward()

    # update
    W.data += -lr * W.grad
print(f'\naverage negative log_likelihood = {loss_tr.item():.4f}\n')

print(f"train_loss (final): {losses_tr[-1]:.4f}")
plt.plot(losses_tr)
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("Loss on training data")
plt.grid() 
plt.plot()
````

# 3. Tune Regularization/Smoothing with Validation Set
-------

````python
#reg_param_list = [0, 0.001, 0.01, 0.02, 0.05, 0.1, 0.25, 0.5, 1.0]
#reg_param_list = [0.5, 0.375, 0.25, 0.1, 0.05, 0.01, 0.005, 0.001, 0.0005, 0.0001, 0.00005, 0.00001][::-1]
#reg_param_list = [0, 0.02, 0.05, 0.1]
reg_param_list = np.geomspace(0.001, 0.5, 12)

loss_dict = {} # <reg_strength: [final train loss, final val loss]>
n_iter = 100
lr = 50

for C in tqdm(reg_param_list):
    # Initialize model & losses
    W = torch.randn((27*2, 27), generator=g, requires_grad=True)
    #tr_losses, val_losses= [], []
    
    # gradient descent
    for k in range(n_iter):

        # forward pass
        xenc = F.one_hot(xs_tr, num_classes=27).float()       
        logits = xenc.view(-1, 27*2) @ W                 
        counts = logits.exp()                             
        probs = counts / counts.sum(1, keepdim=True)     
        tr_loss = -probs[torch.arange(num_tr), ys_tr].log().mean() + C*(W**2).mean() 
        #print(loss.item())
        #tr_losses.append(tr_loss)

        # Zero grad & backward pass
        W.grad = None 
        tr_loss.backward()

        # update
        W.data += -lr * W.grad     

    for w in val:
        chs = ['.'] + list(w) + ['.']
        for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
            ix1 = stoi[ch1]
            ix2 = stoi[ch2]
            ix3 = stoi[ch3]
            xs_val.append((ix1, ix2))
            ys_val.append(ix3)

    xs_val2 = torch.tensor(xs_val) 
    ys_val2 = torch.tensor(ys_val)
    num_val = ys_val2.nelement()

    xenc_val = F.one_hot(xs_val2, num_classes=27).float()      
    logits = xenc_val.view(-1, 27*2) @ W                   
    counts = logits.exp()                             
    probs = counts / counts.sum(1, keepdim=True)     
    val_loss = -probs[torch.arange(num_val), ys_val2].log().mean()

    print(f'For regularization parameter = {C:.4f} ---> train loss: {tr_loss.item():.4f},\
    validation loss: {val_loss.item():.4f}')
    
    _tr_loss = np.round(tr_loss.item(), decimals=4)
    _val_loss = np.round(val_loss.item(), decimals=4)
    _C = np.round(C, decimals=4)
    loss_dict[_C] = [_tr_loss, _val_loss]
````

````python
logits = x_enc.view(-1, 2*27) @ W
counts = logits.exp()
probs = counts / counts.sum(1, keepdim=True)
loss = -probs[torch.arange(x_enc.shape[0]), ys].log().mean()
print(f"Average Negative Log Likelihood Loss: {loss.item():.4f}")
````

````python
loss_ce = F.cross_entropy(logits, ys)
print(f"Cross Entropy Loss: {loss_ce.item():.4f}")
````

We get the same results using both methods. However, `F.cross_entropy` is preferable because it combines the softmax step and loss function step together at once. Basically, `F.cross_entropy` incorporates transforming the neural net outputs to probabilities (1) with calculating the loss in one step.(**`logits`** -> **`losses`**)


- <u>NN outputs to probabilities</u>: **`logits`** -> [exp(x)] -> **`counts`** -> [normalize(x)] -> **`probs`**
- <u>Probabilities to losses</u>: **`probs`** -> [Average negative log likelihood(x)] -> **`losses`**

# 6. meta-exercise: PyTorch Base Class for all Neural Network Modules
-------

````python
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        #print(ch1, ch2, ch3)
        xs.append([ix1, ix2])
        ys.append(ix3)

xs = torch.tensor(xs) 
ys = torch.tensor(ys)
````

````python
class MLP(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = torch.nn.Linear(2*27, 27) # first fully connected layer of MLP
        torch.nn.init.normal_(self.fc1.weight, mean=0, std=1)
        #torch.nn.init.xavier_uniform(self.fc1.weight)

    def forward(self, x):
        logits = W[x[:, 0]] + W[x[:, 1] + 27]
        return logits
````

````python
C = 0
model = MLP()
# gradient descent
for k in range(250):

    # forward pass
    logits = model(xs)
    loss = F.cross_entropy(logits, ys) 
    loss += C * (model.fc1.weight ** 2).mean() # added regularization term
    
#     if k % 10 == 0:
#         print(f'After {k} iterations, the loss is {loss.item():.4f}')

    # backward pass
    model.fc1.zero_grad() # set to zero the gradient (reset gradients of the layer)
    loss.backward()

    # update
    model.fc1.weight.data += -50 * model.fc1.weight.grad
print(f"After {k} iterations, the loss is {loss.item():.4f}")
````
