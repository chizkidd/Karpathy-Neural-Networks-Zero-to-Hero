```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initial learning rate and decay factor
lri = 0.01
k = 10**-8
max_steps = 100000

# List to store the learning rate values
lr_values = []

# Update and collect learning rate values
for i in range(max_steps):
    lr = lri * np.exp(-k * i)
    lr_values.append(lr)
    if i % 10000 == 0:
        print(f"Iteration {i}: Learning Rate = {lr}")

# Plotting the learning rate
plt.figure(figsize=(10, 6))
plt.plot(lr_values, label="Learning Rate")
plt.xlabel("Iteration")
plt.ylabel("Learning Rate")
plt.title("Learning Rate Over Iterations")
# plt.yscale('log')  # Use a log scale for better visualization
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Initial learning rate and decay factor
lri = 0.1
k = 10**-8
max_steps = 100000

# List to store the learning rate values
lr_values = []

# Update and collect learning rate values
for i in range(max_steps):
    lr = lri / (1 + k * i)
    lr_values.append(lr)
    if i % 10000 == 0:
        print(f"Iteration {i}: Learning Rate = {lr}")

# Plotting the learning rate
plt.figure(figsize=(10, 6))
plt.plot(lr_values, label="Learning Rate")
plt.xlabel("Iteration")
plt.ylabel("Learning Rate")
plt.title("Learning Rate Over Iterations")
# plt.yscale('log')  # Use a log scale for better visualization
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define the initial and final learning rates
initial_lr = 0.1
final_lr = 0.01
max_steps = 100000

# Calculate the amount to decrease the learning rate at each step
decay_per_step = (initial_lr - final_lr) / max_steps

# List to store the learning rate values
lr_values = []

# Update and collect learning rate values
for i in range(max_steps+1):
    lr = initial_lr - decay_per_step * i
    lr_values.append(lr)
    if i % 10000 == 0:
        print(f"Iteration {i}: Learning Rate = {lr}")

# Plotting the learning rate
plt.figure(figsize=(10, 6))
plt.plot(lr_values, label="Learning Rate")
plt.xlabel("Iteration")
plt.ylabel("Learning Rate")
plt.title("Linear Decrease of Learning Rate")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Initial learning rate and decay parameters
initial_lr = 0.001
decay_rate = 1e-4
max_steps = 100000

# List to store the learning rate values
lr_values = []

# Update and collect learning rate values
for i in range(max_steps):
    lr = initial_lr / (1 + decay_rate * i)
    lr_values.append(lr)
    if i % 10000 == 0:
        print(f"Iteration {i}: Learning Rate = {lr}")

# Plotting the learning rate
plt.figure(figsize=(10, 6))
plt.plot(lr_values, label="Learning Rate")
plt.xlabel("Iteration")
plt.ylabel("Learning Rate")
plt.title("Learning Rate Over Iterations (Concave)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Initial learning rate and decay parameters
initial_lr = 1e-2  # Lower initial learning rate
decay_rate = 1e-4


# Warmup parameters
warmup_steps = int(max_steps * 0.05)  # Number of warmup steps = 5000
warmup_lr = initial_lr * 10  # Target learning rate after warmup = 0.01

# List to store the learning rate values
lossi = []

# Update and collect learning rate values
for i in range(max_steps):
    # Calculate the learning rate for the current step
    if i < warmup_steps:
        # Perform warmup
        lr = initial_lr + (warmup_lr - initial_lr) * (i / warmup_steps)
    else:
        # Apply inverse decay after warmup
        lr = warmup_lr / (1 + decay_rate * (i - warmup_steps))

    lossi.append(lr)

    if i % 10000 == 0:
        print(f"Iteration {i}: Learning Rate = {lr:.6f}")
        
        
        
        
# ----------------------------------------------------------------------------------------------------------
# Plotting the learning rate
plt.figure(figsize=(10, 6))
iterations = range(max_steps)

# Warmup steps
warmup_iterations = [i for i in range(warmup_steps)]
warmup_lr_ = []
for i in warmup_iterations:
    warmup_lr_.append(initial_lr + (warmup_lr - initial_lr) * (i / warmup_steps))

# After warmup
after_warmup_iterations = [i for i in range(warmup_steps, max_steps)]
after_warmup_lr = []
for i in after_warmup_iterations:
    decay_factor = 1 + decay_rate * (i - warmup_steps)
    after_warmup_lr.append(warmup_lr / decay_factor)

# Plot the learning rate
plt.plot(warmup_iterations, warmup_lr_, label="Warmup")
plt.plot(after_warmup_iterations, after_warmup_lr, label="After Warmup (Inverse Decay)")
plt.xlabel("Iteration")
plt.ylabel("Learning Rate")
plt.title("Learning Rate Over Iterations (Warmup and Inverse Decay)")
plt.grid(True)
plt.legend()
plt.show()
```
